## Chapter 1: Fundamentals

### Offline(Batch) Learning

 - Type of learning that uses up all the available data.
 - A disadvantage for using all the data at once can prevent a system from learning incrementally.
 - The typical process 
    - System is trained and launched into production 
    - When launched the system can't learn anymore and just applies what is has learned. 


### Online Learning 

- Train a system incrementally with mini-batches 
    - mini-batches are samples from a large dataset 
- Normal implementations are for scenarios for adapting quick (Stock market models)
- Learning rate parameter
    - Tells the system how fast to adapt to changing data 
    - A higher learning rate means the system will learn the new data more, which leads to quickly forgetting the previous data.
    - A smaller learning rate tells the system to learn slower. This allows the system to not be as sensitive to new data and outliers.
- For online learning, if bad data is given to the system, the system's performance will gradually decline.

### Categorizing ML Systems

#### Instance Based 

- The machine learns by storing all instances of data and then using these instances to make predictions on new data. 
- The machine compares the new data to the instances it has seen before and uses the closest match to make a prediction.
    - measure of similarity and no model is created 
- Used in pattern recognition, clustering, and anomaly detection. 
- Disadvantages include slower results, less accuracy since it can have a smaller dataset, and limited understanding of data. 


#### Model Based Learning 

- A system uses the dataset to create a model that can predict outcomes based on input data.
- The model is typically created using statistical algorithms such as linear regression, logistic regression, decision trees, and neural networks.
- A disadvantage is the requirement of a large dataset.
